```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(-2, 4, 100)  # x values from -2 to 4
y = x**3 - 3*x**2 + 11*x     # Corresponding y values for the function

# Plot the data
plt.plot(x, y, label="f(x) = x^3 - 3x^2 + 11x", color='blue')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^3 - 3x^2 + 11x')

# Show a legend and the grid
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
