```julia
using Plots

# Generate data
x = range(-2, 4, length=100)  # x values from -2 to 4
y = x.^3 .- 3*x.^2 .+ 11*x    # Corresponding y values for the function

# Plot the data
plot(x, y, label="f(x) = x^3 - 3x^2 + 11x", title="Plot of f(x)", xlabel="x", ylabel="f(x)", legend=true, fmt = :png)

# Display the plot
```
